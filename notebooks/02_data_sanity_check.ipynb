# Checking Data after Preprocessing

Some sanity checks to ensure that the preprocessing is done correctly.

## Load the preprocessed data

```python
# Import Statements
import os
import numpy as np
import pandas as pd
import scipy.io
import matplotlib.pyplot as plt
```

```python
# Load a preprocessed data file

data = np.load("../data/processed/part_1_ppg_only.npz")
X = data["X"]
y = data["y"]

print("X shape:", X.shape)
print("y shape:", y.shape)
```

## Visualize some samples

```python
# PLot some random samples

fig, axs = plt.subplots(3, 3, figsize=(18, 8))

for i in range(3):
    for j in range(3):
        idx = np.random.randint(0, X.shape[0])
        axs[i, j].plot(X[idx])
        axs[i, j].set_title(f"Sample {idx} - SBP: {y[idx,0]:.1f}, DBP: {y[idx,1]:.1f}")
        axs[i, j].set_xlabel("Time (samples)")
        axs[i, j].set_ylabel("PPG Amplitude")
plt.tight_layout()
plt.show()
```

## Feature checks

### SBP and DBP

```python
# PLot histograms of SBP and DBP
plt.figure(figsize=(12, 5))
plt.hist(y[:, 0], bins=30, alpha=0.7, label='Systolic Blood Pressure (SBP)', color='r')
plt.axvline(np.mean(y[:, 0]), color='r', linestyle='dashed', linewidth=1)
plt.hist(y[:, 1], bins=30, alpha=0.7, label='Diastolic Blood Pressure (DBP)', color='g')
plt.axvline(np.mean(y[:, 1]), color='g', linestyle='dashed', linewidth=1)
plt.title('Histogram of Systolic and Diastolic Blood Pressure Values')
plt.xlabel('Blood Pressure (mmHg)')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
